```python
from transformers import AutoTokenizer
import transformers
import torch
```

```python
model = "Zen1t/WordWeaver-AI"
prompt = "В районах с частыми перебоями в подаче электроэнергии рекомендуется выбирать теплосчетчики с автономными источниками питания. Это поможет избежать потери важных данных о потреблении тепла и обеспечит непрерывность работы прибора даже в случае отключения электричества."
prompt = f"<s>[INST] {prompt} [/INST]"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

```python
sequences = pipeline(
    prompt,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    max_length=500,
)
```

```python
for seq in sequences:
    print(seq['generated_text'][len(prompt):])
```
